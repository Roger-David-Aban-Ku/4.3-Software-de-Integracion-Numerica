## METODO DE SIMPSON - CASO GENERAL

* Caso general o compuesta, $h=(b-a)/n$, n par.  $\int_{a}^{b} f(x) dx = \frac{h}{3} [f(a)+ 2 \sum_{j=1}^{(n/2)-1} f(x_{2j}) + 4 \sum_{j=1}^{(n/2)} f(x_{2j-1}) + f(b)]$
$$\int_{0}^{3} e^{2x} \sin (3x) dx$$
$$n=8$$

```python
import math
## definir una funciÃ³n que realice las evaluaciones
def fxi(x):
    return (math.e**(2*x)) * math.sin(3*x)
```

```python
def sumapares(afxi):
    s = len(afxi)
    sumapar= sum([afxi[i] for i in  range(2,s-1) if i%2==0])
    return sumapar

def sumaimpares(afxi):
    s =  len(afxi)
    sumaimpar = sum([afxi[i] for i in  range(1,s-1) if i%2!=0])
    return sumaimpar

def Simpson(afxi,h):
    #[operaciones for  if condicion]
    #sum([])=>suma o un solo valor, escalar, 
    s = len(afxi)
    resultado = (h/3) * (afxi[0]+afxi[s-1] + 2 * sumapares(afxi) + 4* sumaimpares(afxi))
    return resultado
```

```python
def principal():
    n = int(input("Cantidad de intervalos: "))
    a = float(input("Limite a: "))
    b = float(input("Limite b: "))
    h= (b-a)/n
    rangopar=int(n/2 - 1)
    rangoimpar=int(n/2)
    ## Comprension and generator
    afxi = [fxi(a+i*h) for i in range(n+1)] #comprenssion
    resultadosimpson = Simpson(afxi,h)
    print("\033[1;32m""\tMETODO DE SIMPSON")
    print("-----------------------------------------------------------------")
    print("\033[1;30m""\t |i|", "\t |xi|", "\t |fxi|")
    for i in range (n+1):
        print("\t ",i, "\t",(a+i*h), "\t",afxi[i])
    print("-----------------------------------------------------------------")
    print("-----------------------------------------------------------------")
    print("Resultado del primer y ultimo intervalo")
    print("\n\t |i|","\t |fxi|")
    primero=afxi[0]
    ultimo=afxi[n] 
    print("\t ",0, "\t", primero)
    print("\t ",n, "\t", ultimo)
    sumado=primero+ultimo
    print("\033[1;31m""\n**La suma de estos 2 intervalos: ", sumado)
    print("-----------------------------------------------------------------")
    print("-----------------------------------------------------------------")
    comodinsum=0
    for i in range(1):
        print("\033[1;30m""Resultados de los pares de ", i+2, " hasta ", n-2)
        print("\n\t |i|","\t |fxi|")
    for i in range(1, n-1):
        if (i%2==0):
            print("\t ",i, "\t", afxi[i])
            comodinsum+=afxi[i]
    par=2*sumapares(afxi)
    print("\n** Sumatoria de resultado pares: ", comodinsum)
    print("\033[1;32m""** Resultado de la suma ",comodinsum," * 2: ", par)
    print("-----------------------------------------------------------------")
    print("-----------------------------------------------------------------")
    comodinsum2=0
    for i in range(1):
        print("\033[1;30m""Resultados de los impares de ", i+1, " hasta ", n-1)
        print("\n\t |i|","\t |fxi|")
    for i in range(1, n):
        if (i%2!=0):
            print("\t ",i, "\t", afxi[i])
            comodinsum2+=afxi[i]
    impar=4*sumaimpares(afxi)
    print("\n** Sumatoria de resultado impares: ", comodinsum2)
    print("\033[1;35m""** Resultado de la suma", comodinsum2," * 4: ", impar)
    print("-----------------------------------------------------------------")
    print("-----------------------------------------------------------------")
    sumatotal=sumado+par+impar
    print("\033[1;30m""\nSumatoria de""\033[1;31m",sumado,"+""\033[1;32m",par,"+""\033[1;35m",impar)
    print("\033[1;30m""Sumatoria Total: ", sumatotal)
    print("\033[1;34m""Resultado con el Metodo Simpson: ", resultadosimpson)
```
